```python
q1 = rfm['Monetary'].quantile(0.05)
q3 = rfm['Monetary'].quantile(0.95)
IQR = q3-q1
rfm_new= rfm[(rfm['Monetary']>=q1) & (rfm['Monetary']<=q3)]

q_1 = rfm['Frequency'].quantile(0.05)
q_3 = rfm['Frequency'].quantile(0.95)
rfm_new = rfm[(rfm['Frequency']>=q_1) & (rfm['Frequency']<=q_3)]

qt1 = rfm['Recency'].quantile(0.05)
qt3 = rfm['Recency'].quantile(0.95)
rfm_new = rfm[(rfm['Recency']>=qt1) & (rfm['Recency']<=qt3)]

rfm_new
rfm_without_outlier = rfm_new
rfm_without_outlier
```

```python
def group(Q1,Q3,Q4,value):
    if value<=Q1:
        return '1'
    if Q1< value <=Q3:
        return '2'
    if Q3<value<=Q4:
        return '3'
    if Q4<value:
        return '4'
    
```

```python
rec_q1,rec_q2,rec_q3 = rfm['Recency'].quantile([0.25,.050,0.75])
```

```python
recency_group =[]
for i in range(len(rfm)):
    rece = rfm.iloc[i]['Recency']
    recency_group.append(group(rec_q1,rec_q2,rec_q3,rece))

rfm['recency_group'] = recency_group
    
```

```python
freq_q1,freq_q2,freq_q3= rfm['Frequency'].quantile([0.25,0.50,0.75])
```

```python
freq_group = []
for i in range(len(rfm)):
    freq = rfm.iloc[i]['Frequency']
    freq_group.append(group(freq_q1,freq_q2,freq_q3,freq))
rfm['freq_group']= freq_group
```

```python
mon_q1,mon_q2,mon_q3 = rfm['Monetary'].quantile([0.25,0.50,0.75])
```

```python
mon_group=[]
for i in range(len(rfm)):
    mon= rfm.iloc[i]['Monetary']
    mon_group.append(group(mon_q1,mon_q2,mon_q3,mon))
rfm['mon_group']=mon_group
```

```python
rfm
```

```python
rfm['score']=rfm['recency_group'].astype(str)+rfm['freq_group'].astype(str)+rfm['mon_group'].astype(str)
```

```python
rfm['final score']= rfm['recency_group'].astype(int)+rfm['freq_group'].astype(int)+rfm['mon_group'].astype(int)
```

```python
rfm
```

```python
customer_segment = ['Platinum','Diamond','Gold','Silver']
sgmt= pd.qcut(rfm['final score'], q=4, labels = customer_segment)
rfm['Segment']=sgmt.values
rfm
```

```python
def seg_group(score):
    if score =='144':
        return 'Best Customer'
    elif score =='411':
        return 'Beginner'
    elif score  =='444':
        return 'Lost Big Spender'
    elif score  == '441':
        return 'Lost Cheap'
    elif score[1] =='4':
        return 'Loyal'
    elif score[2] =='4':
        return 'Big Spender'
    else :
        return 'Normal Customer'
```

```python
rfm['c_segment']= rfm.score.apply(seg_group)
```
